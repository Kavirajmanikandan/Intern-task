### Data Cleaning and Preprocessing

### Analysis of Successful Teams and Players

```python
# Most Successful team based on the number of wins

team_wins = ipl_data2['winner'].value_counts()
display(team_wins)
```

```python
# Visualize the number of wins by team

plt.figure(figsize = (12, 6))
sns.countplot(x = 'winner', data = ipl_data2)
plt.xticks(rotation = 90)
plt.title('Number of wins by team')
plt.show()
```

```python
# Most successful players based on the number of player of the Match awards

player_of_match = ipl_data2['player_of_match'].value_counts().head(10)
display(player_of_match)
```

```python
# Visualize the top players with the most player of the Match awards

plt.figure(figsize = (12, 6))
sns.countplot(x = 'player_of_match', data = ipl_data2, order = player_of_match.index)
plt.xticks(rotation = 90)
plt.title('Top Players with Most Player of the match Awards')
plt.show()
```

### Factors  Contributing  to  win  or  loss

```python
# Toss decisions and their impact on match outcomes

toss_decision = ipl_data2['toss_decision'].value_counts()
display(toss_decision)
```

```python
# Visualize the impact of toss decision on match outcomes

plt.figure(figsize = (8, 6))
sns.countplot(x = 'toss_decision', hue = 'result', data = ipl_data2)
plt.title('Impact of Toss Decision on Match Outcome')
plt.show()
```

```python
# Venue and its impact on match outcomes

venue_wins = ipl_data2['venue'].value_counts().head(10)
display(venue_wins)
```

```python
# Visualize the impact of venue on match outcomes

plt.figure(figsize = (12, 6))
sns.countplot(x = 'venue', hue = 'result', data = ipl_data2, order = venue_wins.index)
plt.xticks(rotation = 90)
plt.title('Impact of Venues on Match outcomes')
plt.show()
```
