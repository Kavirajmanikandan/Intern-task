### Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```


### Load the Dataset (SampleSuperstore.csv)

```python
df = pd.read_csv('SampleSuperstore.csv')
```

```python
# Display the Dataset(SampleSuperstore.csv)

display(df)
```


### Explore the Data 

```python
#Display the first few rows of the dataset

display(df.head())
```

```python
#Get information about the dataset

display(df.info())
```

```python
#Summary statistics of numerical columns

display(df.describe())
```

```python
#Check for missing values

display(df.isnull().sum())
```

```python
#Check for duplicate rows

display(df.duplicated().sum())
```

```python
#Check unique values in categorical columns

display(df['Category'].unique())
```

```python
#Check unique values in categorical columns

display(df['Sub-Category'].unique())
```

### Data Visualization

```python
#Profit by Category

sns.barplot(x = 'Category', y = 'Profit', data=df)
plt.title('Profit by Category')
plt.show()
```

```python
#Profit by Sub-Category
plt.figure(figsize = (12,6))
sns.barplot(x ='Sub-Category', y='Profit', data=df)
plt.title('Profit by Sub-Category')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Sales by Region

plt.figure(figsize = (12, 6))
sns.barplot(x = 'Region', y='Sales', data = df, estimator = sum)
plt.title('Sales by Region')
plt.show()
```

```python
#Profit by Region

plt.figure(figsize = (12, 6))
sns.barplot(x = 'Region', y='Profit', data = df, estimator = sum)
plt.title('Profit by Region')
plt.show()
```

```python
#Correlation between numerical variables

corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#scatter plot showing sales and profit by state

sns.scatterplot(x="Sales", y="Profit", hue="State", data=df)
plt.title("Sales and Profit by State")
plt.show()
```
